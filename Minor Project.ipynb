```python
import numpy as np 
import pandas as pd
```

```python
superstore_df=pd.read_csv(r'C:\Users\aksha\OneDrive\Desktop\superstore_US.csv')
```

```python
superstore_df=superstore_df.sort_values('State',ascending=True)
```

```python
superstore_df.State
```

```python
superstore_df=superstore_df.rename(columns={'Row ID':'Row_Id','Order ID':'Order_Id',
                                        'Order Date':'Order_date','Ship Date':'Ship_date','Ship Mode':'Ship_mode',
                                           'Customer ID':'Customer_Id','Customer Name':'Customer_name','Segment':'Segment',
                                           'City':'City','State':'State','Country':'Country','Postal Code':'Postal_code',
                                           'Market':'Market','Region':'Region','Product ID':'Product_Id','Category':'Category',
                                           'Sub-Category':'Sub_category','Product Name':'Product_name','Sales': 'Sales',
                                            'Quantity':'Quantity','Discount':'Discount','Profit':'Profit'})
```

```python
superstore_df.columns
```

```python
superstore_new_df=superstore_df.drop(['Row_Id','Customer_name','Customer_Id','Product_Id','Product_name','Country'], axis=1)
```

```python
superstore_new_df.Segment.unique()
print("Unique Segments in this dataset are {}".format(superstore_new_df.Segment.unique()))
```

```python
superstore_new_df.Category.unique()
print("Unique Categories in this dataset are {}".format(superstore_new_df.Category.unique()))
```

```python
superstore_new_df['Order_date'] = pd.to_datetime(superstore_new_df['Order_date']).dt.date
superstore_new_df.Order_date.min()
superstore_new_df.Order_date.max()
print('Order timeframe is from {} to {}' .format(superstore_new_df.Order_date.min(),superstore_new_df.Order_date.max()))
                                              
# Shipdate
superstore_new_df['Ship_date'] = pd.to_datetime(superstore_new_df['Ship_date']).dt.date
superstore_new_df.Ship_date.min()
superstore_new_df.Ship_date.max()
print('Shipping data is from {} to {}' .format(superstore_new_df.Ship_date.min(),superstore_new_df.Ship_date.max()));
```

```python
superstore_new_df['Year']= pd.DatetimeIndex(superstore_new_df.Order_date).year
print("Years included in this analysis are {}" .format(superstore_new_df.Year.unique()))
```

```python
superstore_new_df.isna().sum()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
superstore_new_df.groupby('Category')[['Sales','Profit']].agg('sum')
```

```python
plt.figure(figsize=(12,8))
%matplotlib inline
superstore_new_df.groupby('Category')[['Sales','Profit']].agg('sum').plot.bar(color=["green","blue"])                                                                          
plt.title('Total Profit and Sales per Category')
plt.show();
```

```python
plt.figure(figsize=(15,8))
%matplotlib inline
plt.title("Region wise Purchase Count of each Category")
sns.countplot(x="Category", hue="Region", data=superstore_new_df)
plt.xticks(rotation=90)
plt.show();
```

```python

sub_category_count=superstore_new_df['Sub_category'].value_counts()
sub_category_count
```

```python
%matplotlib inline
plt.figure(figsize=(12,8))
sub_category_count.plot(kind='pie',autopct="%1.1f%%");
```

```python
plt.figure(figsize=(15,8))
plt.title("Region wise Purchase Count of each Sub Category")
sns.countplot(x="Sub_category", hue="Region", data=superstore_new_df)
plt.xticks(rotation=90)
plt.show();
```

```python
plt.figure(figsize=(15,8))
plt.title("Region wise Purchase Count of each State")
sns.countplot(x="Sub_category", hue="State", data=superstore_new_df,width=2.5,)
plt.xticks(rotation=90)
plt.show();
```

```python
sns.countplot(x='Ship_mode',hue='Year',data=superstore_new_df)
```
